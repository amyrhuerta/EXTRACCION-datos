```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.head()
```

```python
df.info()
```

```python
#SOLO QUEREMOS CONOCER EL CLIENTE PUBLICO EN GENERAL
filtro1=df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro1
```

```python
filtro1.to_csv("FILTRO1.csv")
files.download("FILTRO1.csv")
```

```python
# SOLO QUEREMOS CONOCER LOS CLIENTES QUE SE ENCUENTRAN EN LA FILA 523, 1000, 1416
#filtro2= df.iloc[0:21,: ]           #Filas consecutivas
filtro2=df.iloc[ [523, 1000, 1416] , : ]  #Filas no consecutivas 
filtro2
```

```python
filtro2.to_csv("FILTRO2.csv")
files.download("FILTRO2.csv")
```

```python
# DEJAMOS SOLO LAS COLUMNAS DEL NOM DEL CLIENTE, LA CANTIDAD QUE COMPRO Y EL MARGEN UNITARIO
#filtro3=df.iloc[ : , 1:3]       #Columnas consecutivas 
filtro3=df.iloc[ : , [3,4,12]]  #Columnas no consecutivas 
filtro3
```

```python
filtro3.to_csv("FILTRO3.csv")
files.download("FILTRO3.csv")
```

```python
#CREAMOS UN DF CON INDEX DEL NOMBRE DEL CLIENTE
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
df1.head()
```

```python
# VISUALIZAMOS SOLO LOS CLIETES PUBLICO EN GENERAL Y JUAN ANTONIO RESENDIZ CON SU RESPECTIVO MARGEN TOTAL DE CADA COMPRA
filtro4=df1.loc[["PÚBLICO EN GENERAL","JUAN ANTONIO RESENDIZ UGALDE"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
filtro4.to_csv("FILTRO4.csv")
files.download("FILTRO4.csv")
```

```python
#FILTRAMOS SOLO LAS PRIMERAS 10 FILAS
filtro5=df1.head(10)
filtro5
```

```python
filtro5.to_csv("FILTRO5.csv")
files.download("FILTRO5.csv")
```

```python
# FILTRAMOS LAS VENTAS QUE HAN TENIDO MÁS DE 10 UNIDADES
filtro6=df1[df1["CANT"] > 10]
filtro6
```

```python
filtro6.to_csv("FILTRO6.csv")
files.download("FILTRO6.csv")
```

```python
# FILTRAMOS LAS VENTAS QUE TIENEN UN COSTO MENOR Y QUE SE VENDEN EN GRANDES CATIDADES
filtro7=df1[(df1["COSTO_UNITARIO_CALCULADO"] < 300) & (df1["CANT"] >= 20)]
filtro7
```

```python
filtro7.to_csv("FILTRO7.csv")
files.download("FILTRO7.csv")
```

```python
# FILTRAMOS EL TOTAL DE MARGEN SUPERIOR A 26K O VENTAS DEL AÑO EN CURSO 
filtro8=df1[(df1["MARGEN_TOTAL_CALCULADO"] > 26000) | (df1["FECHA_DOC"] >= "2023-01-1")]
filtro8
```

```python
filtro8.to_csv("FILTRO8.csv")
files.download("FILTRO8.csv")
```

```python
# FILTRO DONDE NO CONSIDERE EL CLIENTE PUBLICO EN GENERAL Y QUE NO SEAN VENTAS MAYORES A 28K
filtro9=df[~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL") & ~(df["MARGEN_TOTAL_CALCULADO"] > 28000)]
filtro9
```

```python
filtro9.to_csv("FILTRO9.csv")
files.download("FILTRO9.csv")
```
