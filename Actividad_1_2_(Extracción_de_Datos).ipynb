```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files 
```

```python
df= pd.read_excel('Datos de Facturacion.xlsx')
```

```python
df.info()
```

```python
#CVE_CLPV (valores: 1000 a 2000)
filtro1=df[(df["CVE_CLPV"] >=1000) & (df["CVE_CLPV"] <=12000)]
filtro1
```

```python
filtro1.to_csv("FILTRO1.1.csv")
files.download("FILTRO1.1.csv")
```

```python
# CVE_VEND (Todas las claves excepto “5” y “4”)
filtro2=df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2
```

```python
filtro2.to_csv("FILTRO2.1.csv")
files.download("FILTRO2.1.csv")
```

```python
# FECHA_ENT   (Fechas de “28/02/2022”)
filtro3=df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
```

```python
filtro3.to_csv("FILTRO3.1.csv")
files.download("FILTRO3.1.csv")
```

```python
# CAN_TOT  (Cantidades menores a 5951.7)   o  STATUS (“E”)
filtro4=df[(df["CAN_TOT"] > 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
filtro4.to_csv("FILTRO4.1.csv")
files.download("FILTRO4.1.csv")
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5=df.iloc[ : , [0,6,7,9]]
filtro5
```

```python
filtro5.to_csv("FILTRO5.1.csv")
files.download("FILTRO5.1.csv")
```

```python
# Solo las filas de 7001-7099
filtro6=df.iloc[7001:7099,: ] 
filtro6
```

```python
filtro6.to_csv("FILTRO6.1.csv")
files.download("FILTRO6.1.csv")
```
